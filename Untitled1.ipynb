```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

import scipy as sp
from scipy import constants
```

```python
A = 1.5525
B = .00629
C = .0004

Lambda = np.linspace(100e-9, 1.2e-6, 100)

n_sell = lambda wave_length: np.sqrt(1 + B*wave_length**2/(wave_length**2 - C))
n_cauchy = lambda wave_length: np.sqrt(A + B/wave_length**2 + C/wave_length**4)

fig, ax = plt.subplots()

ax.plot(Lambda, n_sell(Lambda), label='Sellmeier')
ax.plot(Lambda, n_cauchy(Lambda), label='Cauchy')

ax.legend()
```

```python
Lambda = np.linspace(100e-9, 1.2e-6, 500)
omega = np.linspace(1e16, 2.5e16, 500)#2*np.pi*constants.speed_of_light/Lambda

omega_p_Au = 13.8e15
gamma_Au = 1.07e14

alpha_Au = omega_p_Au/np.sqrt(gamma_Au**2 + omega**2)
beta_Au = gamma_Au*alpha_Au**2/(2*omega)

n_R_Au = np.sqrt(
	(1-alpha_Au**2 + np.sqrt((1-alpha_Au**2)**2 + 4*beta_Au**2))/2
)

n_I_Au = beta_Au/n_R_Au

omega_p_Ar = float(14e15)
gamma_Ar = float(3.22e13)

alpha_Ar = omega_p_Au/np.sqrt(gamma_Au**2 + omega**2)
beta_Ar = gamma_Au*alpha_Au**2/(2*omega)

n_R_Ar = np.sqrt(
	(1-alpha_Ar**2 + np.sqrt((1-alpha_Ar**2)**2 + 4*beta_Ar**2))/2
)

n_I_Ar = beta_Ar/n_R_Ar

omega_p_Cu = float(13.4e15)
gamma_Cu = float(1.45e14)

alpha_Cu = omega_p_Cu/np.sqrt(gamma_Cu**2 + omega**2)
beta_Cu = gamma_Cu*alpha_Cu**2/(2*omega)

n_R_Cu = np.sqrt(
	(1-alpha_Cu**2 + np.sqrt((1-alpha_Cu**2)**2 + 4*beta_Cu**2))/2
)

n_I_Cu = beta_Cu/n_R_Cu

fig1, ax1 = plt.subplots()

ax1.plot(omega, n_R_Au, label='n_R Oro')
ax1.plot(omega, n_I_Au, label='n_I Oro')
ax1.plot(omega, n_R_Ar, label='n_R Plata')
ax1.plot(omega, n_I_Ar, label='n_I Plata')
ax1.plot(omega, n_R_Cu, label='n_R Cobre')
ax1.plot(omega, n_I_Cu, label='n_I Cobre')

ax1.legend()
```
