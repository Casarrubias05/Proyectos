```python
import numpy as np
rng = np.random.default_rng()

import matplotlib as mpl
import matplotlib.pyplot as plt

import pandas as pd
```

```python
tabla_angulos = pd.read_excel('AngulosReflexionRefraccion.xlsx', sheet_name=None)
data = tabla_angulos['Cara Plana'][['Ángulo incidencia', 'Ángulo Refracción']]

data_sin = pd.DataFrame({
	'sin_inc': np.sin(np.radians(data['Ángulo incidencia'])),
	'sin_refr': np.sin(np.radians(data['Ángulo Refracción']))
})

x_data, y_data = data_sin['sin_inc'], data_sin['sin_refr']

data_std = np.cos(y_data)*.1

x_data
```

```python
model_prediction = lambda slope, intercept, input_data: slope*np.array(input_data) + intercept

def model_likelihood_log(slope, intercept, input_data, data):
	difference = -(data - model_prediction(slope, intercept, input_data))**2/(2*data_std**2)
	deviations = np.log(2*np.pi*data_std**2)
	#print(np.exp(np.sum(difference - deviations)))
	return np.sum(difference - deviations)

def model_posterior_log(parameters, parameter_limits, input_data, data):

	def model_prior_log(parameter, limits):
		prior = 1/(limits[1] - limits[0]) * int(limits[0] < parameter < limits[1])
		return np.log(prior) if prior else -np.inf

	parameter_priors = np.array([
		model_prior_log(parameters[0], parameter_limits[0]),
		model_prior_log(parameters[1], parameter_limits[1])
	])

	return model_likelihood_log(*parameters, input_data, data)+np.sum(parameter_priors)

def acceptance_pr(parameters_1, parameters_2, para_lims, input_data, data):
	#print(parameters_1)
	update_log = model_posterior_log(parameters_2, para_lims, input_data, data) - model_posterior_log(parameters_1, para_lims, input_data, data)
	return np.exp(min(0, update_log))

slope_lims = [-1, 1]
intercept_lims = [-.5, .5]

trial_slope = rng.normal(np.average(slope_lims), .8)
trial_intercept = rng.normal(np.average(intercept_lims), .3)

model_posterior_log([trial_slope, trial_intercept], [slope_lims, intercept_lims], x_data, y_data)
```

```python
parameters = np.array([.5, 0])[np.newaxis, :]
#parameters = start_parameters
#print(start_parameters[-1], trial_parameters)

for epoch in range(int(5e2)):
	trial_parameters = np.array([rng.normal(parameters[-1, 0], .4), rng.normal(parameters[-1, 1], .1)])
	trial_acceptance = acceptance_pr(parameters[-1], trial_parameters, [slope_lims, intercept_lims], x_data, y_data)
	acceptance_u = rng.uniform()

	#if trial_acceptance > acceptance_u:
	new_parameters = trial_parameters if trial_acceptance > acceptance_u else parameters[-1]
		#print(epoch, 'replacing parameters')
		#print(epoch, parameters)
	#trial_parameters = np.array([rng.normal(parameters[-1, 0], .8), rng.normal(parameters[-1, 1], .3)])
	parameters = np.vstack((parameters, new_parameters))

print(parameters.shape)


fig, ax = plt.subplots(1, 2, figsize=(13, 6))

ax[0].errorbar(x_data, y_data, yerr=data_std, marker='o')
ax[0].plot(x_data, model_prediction(*parameters[-1], y_data))

ax[1].scatter(parameters[:, 0], parameters[:, 1])
```

```python
epoch_likelihood = []
parameters = np.array([.5, 0])[np.newaxis, :]

for epoch in range(int(5e2)):
	trial_parameters = np.array([rng.normal(parameters[-1, 0], .4), rng.normal(parameters[-1, 1], .1)])
	trial_acceptance = acceptance_pr(parameters[-1], trial_parameters, [slope_lims, intercept_lims], x_data, y_data)
	acceptance_u = rng.uniform()

	new_parameters = trial_parameters if trial_acceptance > acceptance_u else parameters[-1]
	parameters = np.vstack((parameters, new_parameters))
	epoch_likelihood.append(np.exp(model_likelihood_log(*parameters[-1], x_data, y_data)))


fig, ax = plt.subplots(1, 2, figsize=(13, 6))

ax[0].errorbar(x_data, y_data, yerr=data_std, marker='o')
ax[0].plot(x_data, model_prediction(*parameters[-1], y_data))
ax[0].set(title='Refracción', xlabel=r'$\sin \theta_i$', ylabel=r'$\sin \theta_t$')

ax[1].scatter(parameters[:, 0], parameters[:, 1])
ax[1].set(xlabel='slope', ylabel='y-intercept')

fig_like, ax_like = plt.subplots()
ax_like.scatter([range(int(5e2))], epoch_likelihood)
```
