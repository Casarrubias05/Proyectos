```python
import sympy as sy
from sympy.abc import c, n, t

import numpy as np

import matplotlib.pyplot as plt
#%matplotlib widget
```

```python
t_np = np.linspace(0, 2*np.pi, 100)

gamma_x = sy.cos(t)/(1+sy.sin(t)**2)
gamma_y = sy.sin(t) * sy.cos(t)/(1+sy.sin(t)**2)

display(gamma_x, gamma_y)
```

```python
gamma_x_np = sy.lambdify(t, gamma_x, 'numpy')
gamma_y_np = sy.lambdify(t, gamma_y, 'numpy')

fig, ax = plt.subplots()
ax.set(xlabel='$x$', ylabel='$y$')

ax.plot(gamma_x_np(t_np), gamma_y_np(t_np))
```
